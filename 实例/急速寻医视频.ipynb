```python
def getResponse(url, param = {}):
    headers = {
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.90 Safari/537.36'
    }
    response = requests.get(url=url, headers=headers, params=param)
    return response
```

```python
# 下载视频和文本
def downloadVideo(audioUrl):
    try:
        #audioUrl, text = getVideoUrlAndText(url)
        if audioUrl == "":
            print('error')
            return
        #text = '\n'.join(text)
        print(audioUrl)
        fileName = (audioUrl.split('/')[-1]).split('.')[0]
        content = getResponse(audioUrl).content
        with open('video/' + fileName + '.mp4', 'wb') as fp:
            fp.write(content)
        print(audioUrl, "successful")
    except Exception as e:
        print(fileName, '  error')
```

```python
downloadVideo('https://www.xunyi.net/uploads/media/201012/7-201012164911.mp4')
```

```python
linkList = []
with open('url.txt', 'r') as fp:
    while True:
        line = fp.readline()
        if not line:
            break
        line = line.replace('\n', '')
        if line.startswith('https'):
            linkList.append(line)
        print(line)
print(len(linkList))
```

```python
for link in linkList:
    print(link)
```

```python
pool = Pool(4)
pool.map(downloadVideo, linkList)
pool.close()
pool.join()
```
