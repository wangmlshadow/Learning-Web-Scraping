```python
import requests
from bs4 import BeautifulSoup
import re
import json
from selenium import webdriver
```

```python
# 增加UA伪装
def getResponse(url):
    headers = {
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36'
    }
    response = requests.get(url=url, headers=headers)
    return response
```

```python
def getHTML(url):
    response = getResponse(url)
    #response.encoding = 'utf-8'
    return response.text
```

```python
# 获取对应视频ID
def getVideoId(url):
    html = getHTML(url)
    res = re.findall(r'<video (.*?)</video>', html)
    res = re.findall(r'class="(.*?)">', str(res))
    if len(res) == 0:
        return ""
    return res[0]
```

```python
# 通过视频页面url获取可以返回json数据的url
def getUrlForGetJson(url):
    id = str(getVideoId(url))
    first = 'https://playvideo.qcloud.com/getplayinfo/v4/1252162195/'
    return first + id
```

```python
# 通过json数据得到视频源地址
def getSrcUrlForVideo(url):
    response = getResponse(url)
    dic_obj = response.json()
    return str(dic_obj['videoInfo'][ 'sourceVideo']['url'])
```

```python
# 通过视频Url下载视频
def downloadVideo(url):
    fileName = getVideoId(url)
    if fileName == "":
        return
    jsonUrl = getUrlForGetJson(url)
    videoUrl = getSrcUrlForVideo(jsonUrl)
    response = getResponse(videoUrl)
    with open("video/" + fileName + ".mp4", 'wb') as fp:
        fp.write(response.content)
    print(fileName, "successful")
```

```python
def getDownloadUrl(url):
    fileName = getVideoId(url)
    if fileName == "":
        return ""
    jsonUrl = getUrlForGetJson(url)
    videoUrl = getSrcUrlForVideo(jsonUrl)
    return videoUrl
    #response = getResponse(videoUrl)
    #with open("video/" + fileName + ".mp4", 'wb') as fp:
        #fp.write(response.content)
    #print(fileName, "successful")
```

```python
def download(videoUrl):
    response = getResponse(videoUrl)
    fileName = videoUrl.split('/')[-2:]
    fileName = '_'.join(fileName)
    with open("video/" + fileName, 'wb') as fp:
        fp.write(response.content)
    print(fileName, "successful")
```

```python
def getPerVideoUrl(url):
    # 获取此页面Url需要用selenium模拟浏览器
    browser = webdriver.Chrome()
    browser.get(url)
    html = browser.page_source
    
    soup = BeautifulSoup(html, 'html.parser')
    
    all_a_href = soup.find_all('a')

    linkLists = []
    for href in all_a_href:
        temp = str(href)
        matchObj = re.match(r'<a href="https://v.familydoctor.com.cn/a/', temp, re.M|re.I)
        if matchObj:
            #temp = str(temp)
            temp = re.sub(r'"><span(.*)', "", temp)
            temp = re.sub(r'<a href="', "", str(temp))
            linkLists.append(temp)
            
    return linkLists
```
