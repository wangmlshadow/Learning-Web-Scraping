```python
def getResponse(url, param = {}):
    headers = {
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.90 Safari/537.36'
    }
    response = requests.get(url=url, headers=headers, params=param)
    return response
```

```python
def getAudioUrlAndText(url):
    page_text = getResponse(url).text
    tree = etree.HTML(page_text)
    # https://files.99.com.cn/audio/202001/20012109420666.mp3
    #audioUrl = 'https:' + tree.xpath('//div[@class="containner_in"]/audio/@src')[0]
    temp = tree.xpath('//audio/@src')
    if len(temp) == 0:
        return "", ""
    audioUrl = 'https:' + temp[0]
    text = tree.xpath('//div[@class="atcle-cont"]/p/text()')
    return audioUrl, text
```

```python
url, text = getAudioUrlAndText('https://www.rxys.com/yinpin/41913.html')
print(url)
print(text)
```

```python
def downloadAudioAndText(url):
    audioUrl, text = getAudioUrlAndText(url)
    if audioUrl == "":
        print('error')
        return
    text = '\n'.join(text)
    fileName = (audioUrl.split('/')[-1]).split('.')[0]
    with open('text/' + fileName + '.txt', 'w', encoding='utf-8') as fp:
        fp.write(text)
    content = getResponse(audioUrl).content
    with open('mp3/' + fileName + '.mp3', 'wb') as fp:
        fp.write(content)
    print(url, "successful")
```

```python
def getUrls(url):
    linkList = []
    page_text = getResponse(url).text
    tree = etree.HTML(page_text)
    audioList = tree.xpath('//div[@class="video_list"]')
    for audio in audioList:
        pageUrl = "https:" + audio.xpath('.//a/@href')[0]
        linkList.append(pageUrl)
    return linkList
```

```python
# 获取视频页面的视频下载地址和对应文本
def getVideoUrlAndText(url):
    response = getResponse(url)
    response.encoding = 'utf-8'
    page_text = response.text
    tree = etree.HTML(page_text)
    temp = tree.xpath('//video/source/@src')    
    if len(temp) == 0:
        return "", ""    
    audioSrcUrl = 'https:' + temp[0]
    text = tree.xpath('//div[@class="atcle-cont"]/p[2]/text()')
    return audioSrcUrl, text
```

```python
url, text = getVideoUrlAndText('https://www.rxys.com/shipin/41465.html')
print(url)
print(text)
```

```python
# 下载视频和文本
def downloadVideoAndText(url):
    audioUrl, text = getVideoUrlAndText(url)
    if audioUrl == "":
        print('error')
        return
    text = '\n'.join(text)
    fileName = (audioUrl.split('/')[-1]).split('.')[0]
    with open('video_text/' + fileName + '.txt', 'w', encoding='utf-8') as fp:
        fp.write(text)
    content = getResponse(audioUrl).content
    with open('video/' + fileName + '.mp4', 'wb') as fp:
        fp.write(content)
    print(url, "successful")
```

```python
downloadVideoAndText('https://www.rxys.com/shipin/41465.html')
```

```python
# 获取每一页的所有视频页链接
def getVideoUrlOfEveryPage(url):
    response = getResponse(url)
    response.encoding = 'utf-8'
    page_text = response.text
    tree = etree.HTML(page_text)
    allPage = tree.xpath('/html/body/div[4]/div/div')
    linkList = []
    for li in allPage:
        pageLink = 'https:' + li.xpath('.//a/@href')[0]
        #print(pageLink)
        linkList.append(pageLink)
    return linkList
```

```python
for link in getVideoUrlOfEveryPage('https://www.rxys.com/shipin/'):
    print(link)
```

```python
for link in getVideoUrlOfEveryPage('https://www.rxys.com/shipin/'):
    print(link)
    downloadVideoAndText(link)
```
