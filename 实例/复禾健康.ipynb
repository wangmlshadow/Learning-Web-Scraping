```python
import requests
from bs4 import BeautifulSoup
import re
import json
from selenium import webdriver
```

```python
# 通过url获取response 仅使用resquests
def getResponse(url):
    headers = {
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.90 Safari/537.36'
    }
    response = requests.get(url=url, headers=headers)
    return response
```

```python
# 获取页面的html文件
def getHTML(url):
    return getResponse(url).text
```

```python
# <audio src="https://file.bohe.cn/fhfile1/M00/05/E1/ooYBAF-FaS6APCXjAAgqcg0-ZSI490.mp3"></audio>
```

```python
# 通过音频源地址爬取音频
def getAudio(url):
    response = getResponse(url)
    path = 'audio/'
    fileName = url.split('/')[-1]
    with open(path + fileName, 'wb') as fp:
        fp.write(response.content)
```

```python
# 获取语音对应的文本
def getAudioText(html):    
    ex = '<p class="dis-txt-info">(.*?)</p>'
    audio_text = re.findall(ex, html, re.S)
    return audio_text[0]
```

```python
# 根据音频播放页面地址获取音频源地址
def getAudioSrcUrl(html):
    ex = '<audio src="(.*?)"></audio>'
    audio_src_url = re.findall(ex, html, re.S)
    return audio_src_url[0]
```

```python
# 从首页获取各个播放页地址
url = 'https://www.fh21.com.cn/voice/list/new/'
html = getHTML(url)
print(html)
```

```python
soup = BeautifulSoup(html, 'html.parser')
all_a_href = soup.find_all('a')

linkLists = []
for href in all_a_href:
    temp = str(href)
    matchObj = re.match(r'<a href="https://www.fh21.com.cn/voice/view/.*html', temp, re.M|re.I)
    if matchObj:
        ex = '<a href="(.*?).html'
        temp = re.findall(ex, temp, re.S)
        linkLists.append(temp[0] + '.html')
linkLists = list(set(linkLists))
```

```python
for link in linkLists:
    print(link)
```

```python
# 获取每一页的音频播放页面链接
def getAllLinkOfEveryPage(url):
    html = getHTML(url)
    soup = BeautifulSoup(html, 'html.parser')
    all_a_href = soup.find_all('a')

    linkLists = []
    for href in all_a_href:
        temp = str(href)
        matchObj = re.match(r'<a href="https://www.fh21.com.cn/voice/view/.*html', temp, re.M|re.I)
        if matchObj:
            ex = '<a href="(.*?).html'
            temp = re.findall(ex, temp, re.S)
            linkLists.append(temp[0] + '.html')
    linkLists = list(set(linkLists))
    return linkLists
```

```python
for link in getAllLinkOfEveryPage('https://www.fh21.com.cn/voice/list/new/'):
    print(link)
```

```python
# 此网站首页通过分页可以得到所有音频播放页
# https://www.fh21.com.cn/voice/list/new/0_0_1.html 到 https://www.fh21.com.cn/voice/list/new/0_0_100.html
# 修改最后的数字即可
```

```python
# 根据相关url爬取数据
def getTextAndAudio(url):
    html = getHTML(url)
    srcUrl = getAudioSrcUrl(html)
    text = getAudioText(html)
    getAudio(srcUrl)
    fileName = srcUrl.split('/')[-1]
    fileName = fileName.replace("mp3", "txt")
    print(fileName)
    # 写入文本数据
    with open('text/' + fileName, 'w', encoding='utf-8') as fp:
        fp.write(text)
    print(fileName, "successful")
```

```python
getTextAndAudio('https://www.fh21.com.cn/voice/view/633100.html')
```

```python
for i in range(1, 101):
    pageUrl = 'https://www.fh21.com.cn/voice/list/new/0_0_' + str(i) + '.html'
    linkLists = getAllLinkOfEveryPage(pageUrl)
    for link in linkLists:
        getTextAndAudio(link)
        print("a audio successful")
    print("a page successful")
```
