```python
import requests
from bs4 import BeautifulSoup
import re
import json
from selenium import webdriver
from lxml import etree
```

```python
def getResponse(url):
    headers = {
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36'
    }
    response = requests.get(url=url, headers=headers)
    return response
```

```python
# 获取每个页面所有视频链接
def getUrlOfEveryPage(url):
    page_text = getResponse(url).text
    tree = etree.HTML(page_text)
    linkList = tree.xpath('/html/body/div[4]/div/a/@href')
    return linkList
```

```python
# 获取每个页面的视频链接
def getDownloadUrl(url):
    page_text = getResponse(url).text
    tree = etree.HTML(page_text)
    srcUrl = tree.xpath('/html/body/div[3]/div[1]//video/source/@src')[0]
    return srcUrl
```

```python
def download(url):
    response = getResponse(url)
    fileName = url.split('/')[-3:]
    fileName = '_'.join(fileName)
    path = 'video/'
    with open(path + fileName, 'wb') as fp:
        fp.write(response.content)
    print(fileName, "   successful download")
```

```python
for i in range(1, 16):
    url = f'https://www.jiankang.com/myft/list/?p={i}'
    for link in getUrlOfEveryPage(url):
        download(getDownloadUrl(link))
    print(f'page {i} download successful')
```
