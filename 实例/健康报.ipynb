```python
import requests
from lxml import etree
import re
```

```python
def getResponse(url):
    headers = {
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36'
    }
    response = requests.get(url=url, headers=headers)
    return response
```

```python
# 获取每个页面所有视频链接
def getUrlOfEveryPage(url):
    page_text = getResponse(url).text
    tree = etree.HTML(page_text)
    linkList = tree.xpath('/html/body/div[6]/div[1]/ul//h4/a/@href')
    return linkList
```

```python
# 获取每个页面的视频链接
def getDownloadUrl(url):
    response = getResponse(url)
    response.encoding = 'utf-8'
    page_text = response.text
    ex = 'http://wenjian.jkb.com.cn(.*?).mp4'
    temp = re.findall(ex, page_text, re.S)
    if len(temp) == 0:
        return ""
    
    srcUrl = 'http://wenjian.jkb.com.cn' + temp[0] + '.mp4'
    return srcUrl
```

```python
def download(url):
    response = getResponse(url)
    fileName = url.split('/')[-1]
    #fileName = '_'.join(fileName)
    path = 'video/'
    with open(path + fileName, 'wb') as fp:
        fp.write(response.content)
    print(fileName, "   successful download")
```

```python
with open('srcUrl.txt', 'w', encoding='utf-8') as fp:
    for i in range(2, 8):
        url = f'http://www.jkb.com.cn/video/{i}.html'
        for link in getUrlOfEveryPage(url):
            srcUrl = getDownloadUrl(link)
            if srcUrl == "":
                continue
            print(srcUrl)
            #download(srcUrl)
            fp.write(srcUrl)
            fp.write('\n')
        print(f'page {i}  successful')
```

```python
linkList = []
with open('srcUrl.txt') as fp:
    while True:
        line = fp.readline()
        if line:
            line = line.replace('\n', '')
            print (line)            
            linkList.append(line)
        else:
            break
```
