```python
def getResponse(url):
    headers = {
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36'
    }
    response = requests.get(url=url, headers=headers)
    #response.encoding = 'utf-8'
    return response
```

```python
# 下载音频
def download(url):
    # http://cnvod.cnr.cn/audio2017/ondemand/media/1100/201909/5D81E223-6204-4DC7-8CFB-38650A141C1A_2019-09-1815_41_59_0.m4a
    try:
        fileName = url.split('/')[-1]
        path = 'audio/'
        content = getResponse(url).content
        with open(path + fileName, 'wb') as fp:
            fp.write(content)
        print(fileName, "successful")
    except Exception as e:
        print("Exception")
```
