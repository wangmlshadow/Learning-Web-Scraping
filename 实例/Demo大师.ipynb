```python
import requests
from bs4 import BeautifulSoup
import re
import json
from selenium import webdriver
from lxml import etree
import os
from os import path
import shutil
import time
import demjson
from selenium.webdriver.chrome.options import Options
from multiprocessing.dummy import Pool
```

```python
def getResponse(url):
    headers = {
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36'
    }
    response = requests.get(url=url, headers=headers)
    return response
```

```python
# 通过selenium获取page_text
def getHTMLBySelenium(url):
    chrome_options =Options()
    chrome_options.add_argument('--headless')
    browser = webdriver.Chrome(executable_path='C:/Users/Administrator/Desktop/chromedriver_win32/chromedriver.exe', options=chrome_options)
    browser.get(url)
    time.sleep(2)
    page_text = browser.page_source
    browser.quit()
    return page_text
```

```python
def getBlog(url):
    page_text = getResponse(url).text
    tree = etree.HTML(page_text)
    allText = tree.xpath('//div[@class="aw-mod aw-question-detail aw-item"]//text()')
    text = '\n'.join(allText)
    title = tree.xpath('//div[@class="mod-head no-line"]/h1/text()')
    if len(title) == 0:
        return
    title = title[0]
    title = title.replace('/', '_')
    title = title.replace('"', '_')
    title = title.replace('|', '_')
    title = title.replace('?', '_')
    try:
        with open('text/' + title + '.txt', 'w', encoding='utf-8') as fp:
            fp.write(text)
    except Exception as e:
        return
```

```python
def getBlogUrls(page):
    url = f'http://www.demodashi.com/portal/demo/getDemoList4page?queryString=&toPage={page}&orderBy=&type=&typeId=&ptype=&to1=demo&to2=buy&userId='
    blogList = []
    objList = getResponse(url).json()
    
    for obj in objList['data']:
        blogUrl = 'http://www.demodashi.com/demo/' + obj['ID'] + '.html'
        #print(blogUrl)
        blogList.append(blogUrl)        
    return blogList
```

```python
urlSet = set()
```

```python
page = 0
while page < 66:
    print(f'page {page}')
    blogUrlList = getBlogUrls(page)
    page += 1
    for blogUrl in blogUrlList:
        if blogUrl in urlSet:
            continue
        getBlog(blogUrl)
        print(blogUrl)
        urlSet.add(blogUrl)
        with open('url.txt', 'a', encoding='utf-8') as fp:
            fp.write(blogUrl)
            fp.write('\n')
```
