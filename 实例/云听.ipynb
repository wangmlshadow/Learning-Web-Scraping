```python
import requests
from bs4 import BeautifulSoup
import re
import json
from selenium import webdriver
from lxml import etree
import os
from os import path
import shutil
import time
import demjson
from selenium.webdriver.chrome.options import Options
```

```python
def getResponse(url):
    headers = {
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36'
    }
    response = requests.get(url=url, headers=headers)
    #response.encoding = 'utf-8'
    return response
```

```python
def getHTML(url):
    response = getResponse(url)
    response.encoding = 'utf-8'
    return response.text
```

```python
# 获取每一页的音频地址
def getSrcUrl(url):
    chrome_options =Options()
    chrome_options.add_argument('--headless')
    browser = webdriver.Chrome(options=chrome_options)
    browser.get(url)
    page_text = browser.page_source
    tree = etree.HTML(page_text)
    all_tr = tree.xpath('//tbody/tr')
    linkList = []
    for tr in all_tr:
        srcUrl = tr.xpath('./td[@class="td_content"]/a/@data-url')[0]
        #print(srcUrl)
        linkList.append(srcUrl)
    return linkList
```

```python
def getPageUrl(url):
    chrome_options =Options()
    chrome_options.add_argument('--headless')
    browser = webdriver.Chrome(options=chrome_options)
    browser.get(url)
    page_text = browser.page_source
    tree = etree.HTML(page_text)
    all_li = tree.xpath('//*[@id="wrapper"]/li')
    linkList = []
    for li in all_li:
        pageUrl = li.xpath('./a/@href')[0]
        pageUrl = pageUrl.replace('..', 'http://www.radio.cn/pc-portal')
        print(pageUrl)
        linkList.append(pageUrl)
    return linkList
```

```python
# 下载音频
def download(url):
    # http://cnvod.cnr.cn/audio2017/ondemand/media/1100/201909/5D81E223-6204-4DC7-8CFB-38650A141C1A_2019-09-1815_41_59_0.m4a
    try:
        fileName = url.split('/')[-1]
        path = 'audio/'
        content = getResponse(url).content
        with open(path + fileName, 'wb') as fp:
            fp.write(content)
        print(fileName, "successful")
    except Exception as e:
        print("Exception")
```
