```python
import requests
from bs4 import BeautifulSoup
import re
import json
from selenium import webdriver
from lxml import etree
```

```python
def getResponse(url):
    headers = {
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36'
    }
    response = requests.get(url=url, headers=headers)
    return response
```

```python
# 获取每一页所有的视频链接
def getVideoUrlInEveryPage(url):
    page_text = getResponse(url=url).text
    linkList = []
    tree = etree.HTML(page_text)
    all_href = tree.xpath('/html/body/div[2]/div[3]/div[1]/a/@href')
    for href in all_href:
        #print(href)
        linkList.append('https://www.baikemy.com' + str(href))
    return linkList
```

```python
for link in getVideoUrlInEveryPage('https://www.baikemy.com/video/index?page-index=1&page-size=30'):
    print(link)
```

```python
# 获取每个视频源地址
def getSrcUrl(url):
    page_text = getResponse(url=url).text
    tree = etree.HTML(page_text)
    srcUrl = tree.xpath('/html/body/div[2]/div[3]/div[1]/div/video/@src')[0]
    return srcUrl
```

```python
print(getSrcUrl('https://www.baikemy.com/video/details/70000000000013'))
```

```python
def download(url):
    response = getResponse(url)
    fileName = url.split('/')[-1]
    path = 'D:/work/video_2/'
    with open(path + fileName, 'wb') as fp:
        fp.write(response.content)
    print(fileName, "   successful download")
```

```python
download('https://v.baikemy.com/obj/project-l-cdn-video/v/2020-09-15/ganaijiancha.mp4')
```

```python
# 先抓取所有视频下载链接
allDownloadUrl = []
with open('srcUrl.txt', 'w', encoding='utf-8') as fp:
    for i in range(1, 376):
        pageUrl = f'https://www.baikemy.com/video/index?page-index={i}&page-size=30'
        for link in getVideoUrlInEveryPage(pageUrl):
            srcUrl = getSrcUrl(link)
            fp.write(srcUrl)
            fp.write('\n')
            allDownloadUrl.append(srcUrl)
            print(srcUrl)
        print(f'page {i} successful')
```
