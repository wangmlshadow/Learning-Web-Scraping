```python
import requests
from bs4 import BeautifulSoup
import re
import json
from selenium import webdriver
from lxml import etree
import os
from os import path
import shutil
import time
import demjson
```

```python
def getResponse(url, param = {}):
    headers = {
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.90 Safari/537.36'
    }
    response = requests.get(url=url, headers=headers, params=param)
    return response
```

```python
# 获取音频页面的音频下载地址和对应文本
def getAudioUrlAndText(url):
    response = getResponse(url)
    response.encoding = 'utf-8'
    page_text = response.text
    tree = etree.HTML(page_text)
    temp = tree.xpath('//source/@src')    
    if len(temp) == 0:
        return "", ""    
    audioSrcUrl = temp[0]
    text = tree.xpath('//div[@class="video_introduce"]//p/span/text()')
    return audioSrcUrl, text
```

```python
url, text = getAudioUrlAndText('https://www.xinglinpukang.com/audio/31725657.html')
print(url)
print(text)
```

```python
# 下载音频和文本
def downloadAudioAndText(url):
    audioUrl, text = getAudioUrlAndText(url)
    if audioUrl == "":
        print('error')
        return
    text = '\n'.join(text)
    fileName = (audioUrl.split('/')[-1]).split('.')[0]
    with open('text/yjks/' + fileName + '.txt', 'w', encoding='utf-8') as fp:
        fp.write(text)
    content = getResponse(audioUrl).content
    with open('mp3/yjks/' + fileName + '.mp3', 'wb') as fp:
        fp.write(content)
    print(url, "successful")
```

```python
downloadAudioAndText('https://www.xinglinpukang.com/audio/31725657.html')
```

```python
# 获取每一页的所有音频页链接
def getAudioUrlOfEveryPage(url):
    response = getResponse(url)
    response.encoding = 'utf-8'
    page_text = response.text
    tree = etree.HTML(page_text)
    allPage = tree.xpath('//*[@id="audio_list_table"]/div[2]/div[1]/div[1]/a')
    linkList = []
    for li in allPage:
        pageLink = 'https://www.xinglinpukang.com' + li.xpath('./@href')[0]
        #print(pageLink)
        linkList.append(pageLink)
    return linkList
```

```python
# 获取视频页面的视频下载地址和对应文本
def getVideoUrlAndText(url):
    response = getResponse(url)
    response.encoding = 'utf-8'
    page_text = response.text
    tree = etree.HTML(page_text)
    temp = tree.xpath('//source/@src')    
    if len(temp) == 0:
        return "", ""    
    audioSrcUrl = temp[0]
    text = tree.xpath('/html/body/div[1]/div[2]/div[4]/div[2]/div[2]/p/span/text()')
    return audioSrcUrl, text
```

```python
url, text = getVideoUrlAndText('https://www.xinglinpukang.com/vod/66756720.html')
print(url)
print(text)
```

```python
# 下载视频和文本
def downloadVideoAndText(url):
    audioUrl, text = getVideoUrlAndText(url)
    if audioUrl == "":
        print('error')
        return
    text = '\n'.join(text)
    fileName = (audioUrl.split('/')[-1]).split('.')[0]
    with open('video_text/' + fileName + '.txt', 'w', encoding='utf-8') as fp:
        fp.write(text)
    content = getResponse(audioUrl).content
    with open('video/' + fileName + '.mp4', 'wb') as fp:
        fp.write(content)
    print(url, "successful")
```

```python
# 获取每一页的所有视频页链接
def getVideoUrlOfEveryPage(url):
    response = getResponse(url)
    response.encoding = 'utf-8'
    page_text = response.text
    tree = etree.HTML(page_text)
    allPage = tree.xpath('//div[@class="list clears"]/a')
    linkList = []
    for li in allPage:
        pageLink = 'https://www.xinglinpukang.com' + li.xpath('./@href')[0]
        #print(pageLink)
        linkList.append(pageLink)
    return linkList
```
