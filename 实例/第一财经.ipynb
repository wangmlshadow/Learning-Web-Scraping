```python
import requests
from bs4 import BeautifulSoup
import re
import json
from selenium import webdriver
from lxml import etree
import os
from os import path
import shutil
import time
import demjson
from selenium.webdriver.chrome.options import Options
```

```python
def getResponse(url):
    headers = {
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36'
    }
    response = requests.get(url=url, headers=headers)
    #response.encoding = 'utf-8'
    return response
```

```python
url = 'https://www.yicai.com/api/ajax/getjuhelist?action=video&page=5&pagesize=25'
res = getResponse(url)
temp = res.text
jsonList = json.loads(temp)
print(type(jsonList))
print(len(jsonList))
```

```python
# 获取所有链接
def getMoreList(url):
    res = getResponse(url)
    temp = res.text
    jsonList = json.loads(temp)
    linkList = []
    for obj in jsonList:
        pageUrl = 'https://www.yicai.com' + obj['url']
        #print(pageUrl)
        linkList.append(pageUrl)
    return linkList
```

```python
getMoreList('https://www.yicai.com/api/ajax/getjuhelist?action=video&page=1&pagesize=25')
```

```python
# 获取视频地址并下载
def getVideoUrl(url):
    #下面加在setup最上面
    chrome_options =Options()
    chrome_options.add_argument('--headless')
    browser = webdriver.Chrome(options=chrome_options)

    #browser = webdriver.Chrome()
    browser.get(url)
    html = browser.page_source
    tree = etree.HTML(html)
    src = tree.xpath('//video/@src')[0]
    return src
```

```python
print(getVideoUrl('https://www.yicai.com/video/101019021.html'))
```

```python
def download(url):
    content = getResponse(url).content
    fileName = (url.split('?')[0]).split('/')[-1]
    path = 'video/'
    with open(path + fileName, 'wb') as fp:
        fp.write(content)
        print(fileName, "successful")
```

```python
def getAndDownload(url):
    try:
        src = getVideoUrl(url)
        download(src)
    except Exception as e:
        print('error')
```

```python
# https://www.yicai.com/api/ajax/getjuhelist?action=video&page=2&pagesize=25
# https://www.yicai.com/api/ajax/getjuhelist?action=video&page=3&pagesize=25
page = 31
while True:
    print(f'page {page}')
    url = f'https://www.yicai.com/api/ajax/getjuhelist?action=video&page={page}&pagesize=25'
    page = page + 1
    linkList = getMoreList(url)
    if len(linkList) == 0:
        break
    for i, link in enumerate(linkList):
        print(f'video {i}')
        getAndDownload(link)
```
