```python
import requests
from bs4 import BeautifulSoup
import re
import json
from selenium import webdriver
from lxml import etree
import os
from os import path
import shutil
import time
import demjson
from selenium.webdriver.chrome.options import Options
from multiprocessing.dummy import Pool
```

```python
def getResponse(url):
    headers = {
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36'
    }
    response = requests.get(url=url, headers=headers)
    return response
```

```python
# 通过selenium获取page_text
def getHTMLBySelenium(url):
    chrome_options =Options()
    chrome_options.add_argument('--headless')
    browser = webdriver.Chrome(executable_path='C:/Users/Administrator/Desktop/chromedriver_win32/chromedriver.exe', options=chrome_options)
    browser.get(url)
    time.sleep(0.1)
    page_text = browser.page_source
    browser.quit()
    return page_text
```

```python
def getBolg(url):
    page_text = getResponse(url).text
    tree = etree.HTML(page_text)
    allText = tree.xpath('//div[@class="article-box__inner"]//text()')
    text = '\n'.join(allText)
    title = tree.xpath('//h1[@class="article-box__title"]/a/text()')
    if len(title) == 0:
        return
    title = title[0]
    title = title.replace('/', '_')
    title = title.replace('"', '_')
    title = title.replace('|', '_')
    title = title.replace('?', '_')
    try:
        with open('大前端/' + title + '_1.txt', 'w', encoding='utf-8') as fp:
            fp.write(text)
    except Exception as e:
        return
```

```python
def getBlogUrls(page):
    url = f'https://www.oschina.net/group/widgets/_entry_list?groupId=28&circle=&tab=recommend&sort=newest&page={page}&type=ajax'
    blogUrlList = []
    headers = {
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36',
        'Referer':'https://www.oschina.net/group/cross-front'
    }
    response = requests.get(url=url, headers=headers)
    page_text = response.text
    tree = etree.HTML(page_text)
    allNews = tree.xpath('//div[@class="baiduab-beha entry-item entry-item--full entry-item--full-can-expand"]')
    for news in allNews:
        newsUrl = news.xpath('.//a[@class="btn-more btn-more--fixed"]/@href')
        if len(newsUrl) == 0:
            continue
        #print(newsUrl[0])
        blogUrlList.append(newsUrl[0])
    return blogUrlList
```

```python
urlList = []
with open('大前端.txt', 'r', encoding='utf-8') as fp:
    while True:
        line = fp.readline()
        if not line:
            break
        line = line.replace('\n', '')
        print(line)
        urlList.append(line)
print(len(urlList))
urlSet = set(urlList)
print(len(urlSet))
```

```python
page = 773
while True:
    print(f'page {page}')
    blogUrlList = getBlogUrls(page)
    page += 1
    for blogUrl in blogUrlList:
        if blogUrl in urlSet:
            continue
        getBolg(blogUrl)
        print(blogUrl)
        urlSet.add(blogUrl)
        with open('大前端.txt', 'a', encoding='utf-8') as fp:
            fp.write(blogUrl)
            fp.write('\n')
```
