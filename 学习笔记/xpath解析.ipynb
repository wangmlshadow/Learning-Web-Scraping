# xpath解析

```python
# 最常用且最便捷高效的解析方式
# 通用性最强

# xpath解析原理
# 实例化一个etree的对象，并且需要将被解析的页面源码数据加载到该对象中
# 调用etree对象中的xpath方法结合着xpath表达式实现标签定位和内容的捕获

# 如何实例化etree对象
# etree.parse(filePath) 从本机文件实例化
# etree.HTML('page_text') 从互联网获得的html实例化

# xpath('xpath表达式')
# / 斜杠表示从根节点开始定位 
# / 一个斜杠表示一个层级
# // 表示多个层级 可以表示从任意位置开始定位

# 属性定位 tag[@attrName="attrValue"]

# 索引定位 tag[@attrName="attrValue"]/nextAttrName[index] 
# 注意索引是从 1 开始的

# 取文本
# /text()  获取标签中直系的文本内容
# //text() 获取标签中非直系的文本内容 （所有的文本内容）

# 取标签
# /@attrName      img/@src
```

```python
# xpath案例
import requests
from lxml import etree
```

```python
# 爬取到页面源码数据
url = 'https://tj.58.com/ershoufang/'
headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.90 Safari/537.36'
}
page_text = requests.get(url=url, headers=headers).text
```

```python
# 数据解析
tree = etree.HTML(page_text)
all_house = tree.xpath('//div[@class="property"]')
#print(len(all_house))
i = 0
for house in all_house:
    title = house.xpath('.//div[@class="property-content-title"][1]/h3/text()')# . 点表示xpath参照物不再是整个html源码 而是house
    print(title, "asdasd")
    i += 1
print(i)
```

```python
url = 'https://pic.netbian.com/4kdongwu/'
headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.90 Safari/537.36'
}

response = requests.get(url=url, headers=headers)

# 中文乱码问题处理

# 手动设定相应数据的编码格式
print(response.encoding)
response.encoding = 'gbk'

page_text = response.text
print(page_text)
```

```python
import os
if not os.path.exists('./picLibs'):
    os.mkdir('./picLibs')
```

```python
tree = etree.HTML(page_text)
# 数据解析：src的属性值 alt的属性值
li_list = tree.xpath('//div[@class="slist"]/ul/li')# 获取所有li标签
for li in li_list:
    img_src = 'https://pic.netbian.com/' + li.xpath('./a/img/@src')[0]
    img_name = li.xpath('./a/img/@alt')[0] + '.jpg'
    # 通用处理中文乱码的解决方式
    #img_name = img_name.encode('iso-8859-1').decode('gbk')
    #print(img_name, img_src)# 乱码问题
    # 持久化存储
    img_data = requests.get(url=img_src, headers=headers).content
    img_path = 'picLibs/' + img_name
    with open(img_path, 'wb') as fp:
        fp.write(img_data)
        print(img_name, '下载成功。')
```

```python
url = 'https://www.aqistudy.cn/historydata/'
headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.90 Safari/537.36'
}
page_text = requests.get(url=url, headers=headers).text
tree = etree.HTML(page_text)
```

```python
# 热门城市
li_list = tree.xpath('//div[@class="bottom"]/ul/li')
allCityNames = []
for li in li_list:
    hot_city_name = li.xpath('./a/text()')[0]# 城市名称
    allCityNames.append(hot_city_name)    
```

```python
# 所有城市
city_names = tree.xpath('//div[@class="bottom"]/ul/div[2]/li')
for li in city_names:
    city_name = li.xpath('./a/text()')[0]
    allCityNames.append(city_name)
```

```python
for name in allCityNames:
    print(name)
```

```python
url = 'https://www.aqistudy.cn/historydata/'
headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.90 Safari/537.36'
}
page_text = requests.get(url=url, headers=headers).text
tree = etree.HTML(page_text)
allCityNames = []
```

```python
# 解析到热门城市和所有城市对应的a标签
# 层级关系
# div/ul/li/a
# div/ul/div[2]/li/a
a_list = tree.xpath('//dic[@class="bottom"]/ul/li/a | //div[@class="bottom"]/ul/div[2]/li')# 使用 | 使得两个xpath表达式可以同时作用
for a in a_list:
    name = a.xpath('./text()')[0]
    allCityNames.append(name)
```
