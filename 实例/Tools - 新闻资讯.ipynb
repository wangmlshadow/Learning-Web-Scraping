```python
import requests
from bs4 import BeautifulSoup
import re
import json
from selenium import webdriver
from lxml import etree
import os
from os import path
import shutil
import time
import demjson
from selenium.webdriver.chrome.options import Options
from multiprocessing.dummy import Pool
```

```python
def getResponse(url):
    headers = {
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36'
    }
    response = requests.get(url=url, headers=headers)
    return response
```

```python
def getHTMLBySelenium(url):
    try:
        chrome_options =Options()
        chrome_options.add_argument('--headless')
        browser = webdriver.Chrome(executable_path='C:/Users/Administrator/Desktop/chromedriver_win32/chromedriver.exe', options=chrome_options)
        browser.get(url)
        time.sleep(2)
        page_text = browser.page_source
        browser.quit()
        return page_text
    except Exception as e:
        return ''
```

```python
def getBlog(url):   
    try:
        page_text = getResponse(url).text
        tree = etree.HTML(page_text)
        allText = tree.xpath('//div[@class="col main-content col_9_of_12"]//text()')
        text = '\n'.join(allText)
        title = url.replace('/', '_')
        title = title.replace('.', '_')
        title = title.replace(':', '_')
        with open('新闻资讯/' + title + '.txt', 'w', encoding='utf-8') as fp:
            fp.write(text)
    except Exception as e:
        return
```

```python
 def getLinks(url):
    try:
        page_text = getResponse(url).text
        tree = etree.HTML(page_text)
        links = []

        articleList = tree.xpath('//div[@id="articleslist"]/div[@class="col col_12_of_12"]')
        for article in articleList:
            link = article.xpath('.//div[@class="thumb_hover"]/a/@href')
            if len(link) == 0:
                continue
            link = 'https://www.t00ls.net/' + link[0]
            links.append(link)
        return links
    except Exception as e:
        return []
```

```python
page = 388
while True:
    print(f'page {page}')
    url = f'https://www.t00ls.net/news.html?page={page}'
    page += 1
    for link in getLinks(url):
        getBlog(link)
        with open('新闻资讯.txt', 'a', encoding='utf-8') as fp:
            print(link)
            fp.write(link)
            fp.write('\n')
```
