```python
import requests
from bs4 import BeautifulSoup
import re
import json
from selenium import webdriver
from lxml import etree
import os
from os import path
import shutil
import time
import demjson
from selenium.webdriver.chrome.options import Options
from multiprocessing.dummy import Pool
```

```python
def getResponse(url):
    headers = {
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36'
    }
    response = requests.get(url=url, headers=headers)
    return response
```

```python
def getHTMLBySelenium(url):
    try:
        chrome_options =Options()
        chrome_options.add_argument('--headless')
        browser = webdriver.Chrome(executable_path='C:/Users/Administrator/Desktop/chromedriver_win32/chromedriver.exe', options=chrome_options)
        browser.get(url)
        time.sleep(2)
        page_text = browser.page_source
        browser.quit()
        return page_text
    except Exception as e:
        return ''
```

```python
def getBlog(url):   
    try:
        page_text = getResponse(url).text
        tree = etree.HTML(page_text)
        allText = tree.xpath('//div[@id="content"]//text()')
        
        if len(allText) == 0:
            browser = webdriver.Chrome(executable_path='C:/Users/Administrator/Desktop/chromedriver_win32/chromedriver.exe')
            browser.get('https://www.iteye.com/news/32316')
            time.sleep(15)
            browser.quit()
        
        text = '\n'.join(allText)
        title = url.replace('/', '_')
        title = title.replace('.', '_')
        title = title.replace(':', '_')
        with open('问答/' + title + '.txt', 'w', encoding='utf-8') as fp:
            fp.write(text)
    except Exception as e:
        return
```

```python
 def getLinks(url):
    try:
        page_text = getResponse(url).text
        tree = etree.HTML(page_text)
        links = []
        blogList = tree.xpath('//*[@id="ask_list"]/div[@class="question-summary"]')
        for blog in blogList:
            link = blog.xpath('./div[@class="summary"]/h3[1]/a/@href')
            if len(link) == 0:
                continue
            link = 'https://www.iteye.com' + link[0]
            links.append(link)
        return links
    except Exception as e:
        return []
```

```python
page = 28
while page < 5332:
    print(f'page {page}')
    url = f'https://www.iteye.com/problems/hot?page={page}'
    links = getLinks(url)
    if len(links) == 0:
        browser = webdriver.Chrome(executable_path='C:/Users/Administrator/Desktop/chromedriver_win32/chromedriver.exe')
        browser.get('https://www.iteye.com/news/32316')
        time.sleep(10)
        browser.quit()
        continue
    page += 1
    for link in links:
        getBlog(link)
        with open('问答.txt', 'a', encoding='utf-8') as fp:
            print(link)
            fp.write(link)
            fp.write('\n')
        time.sleep(2)
```
