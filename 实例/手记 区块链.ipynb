```python
import requests
from bs4 import BeautifulSoup
import re
import json
from selenium import webdriver
from lxml import etree
import os
from os import path
import shutil
import time
import demjson
from selenium.webdriver.chrome.options import Options
from multiprocessing.dummy import Pool
```

```python
def getResponse(url):
    headers = {
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36'
    }
    response = requests.get(url=url, headers=headers)
    return response
```

```python
# 通过selenium获取page_text
def getHTMLBySelenium(url):
    chrome_options =Options()
    chrome_options.add_argument('--headless')
    browser = webdriver.Chrome(executable_path='C:/Users/Administrator/Desktop/chromedriver_win32/chromedriver.exe', options=chrome_options)
    browser.get(url)
    time.sleep(2)
    page_text = browser.page_source
    browser.quit()
    return page_text
```

```python
def getBlog(url):
    page_txet = getResponse(url).text
    tree = etree.HTML(page_txet)
    title = tree.xpath('//h1[@class="detail-title"]/span/text()')
    allText = tree.xpath('//div[@class="left_essay l"]//text()')
    text = '\n'.join(allText)
    if len(title) == 0:
        return
    title = title[0]
    title = title.replace('/', '_')
    title = title.replace('"', '_')
    title = title.replace('|', '_')
    title = title.replace('?', '_')
    try:
        with open('区块链/' + title + '.txt', 'w', encoding='utf-8') as fp:
            fp.write(text)
    except Exception as e:
        return
```

```python
def getBlogUrls(page):
    url = f'https://www.imooc.com/article/getarticlelist?marking=blockchain&page={page}'
    blogList = []
    page_text = getResponse(url).text
    tree = etree.HTML(page_text)
    liList = tree.xpath('//div[@class="article-lwrap clearfix"]')
    for li in liList:
        blogUrl = li.xpath('./div[@class="list-content l"]/a/@href')
        if len(blogUrl) == 0:
            continue
        blogUrl = 'https://www.imooc.com/' + blogUrl[0]
        #print(blogUrl)
        blogList.append(blogUrl)
    return blogList
```

```python
# 保存已抓取的链接 避免重复抓取
urlList = []
with open('区块链.txt', 'r', encoding='utf-8') as fp:
    while True:
        line = fp.readline()
        if not line:
            break
        line = line.replace('\n', '')
        print(line)
        urlList.append(line)
print(len(urlList))
urlSet = set(urlList)
print(len(urlSet))
```

```python
page = 31
while True:
    print(f'page {page}')
    blogUrlList = getBlogUrls(page)
    page += 1
    for blogUrl in blogUrlList:
        if blogUrl in urlSet:
            continue
        getBlog(blogUrl)
        print(blogUrl)
        urlSet.add(blogUrl)
        with open('区块链.txt', 'a', encoding='utf-8') as fp:
            fp.write(blogUrl)
            fp.write('\n')
```
